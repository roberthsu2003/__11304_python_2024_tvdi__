```python
#建立dict
d1:dict = {"one":1,
           "two":2,
           "three":3}

#檢查元素數量
print(len(d1))

#可以檢查型別
print(type(d1))
```

```python
#subscript[key]
print(d1['one'])
print(d1['two'])
print(d1['three'])
```

```python
#整合for in 迴圈
for key in d1:
    print(f"key={key}")
    print(f"value={d1[key]}")
    print("=================")
```
